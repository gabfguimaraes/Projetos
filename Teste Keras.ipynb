```python
import matplotlib.pyplot as plt
import tensorflow_hub as hub
import tensorflow as tf
import keras
from PIL import Image
import requests
import numpy as np
import io
```

```python
module_handle = 'https://tfhub.dev/google/faster_rcnn/openimages_v4/inception_resnet_v2/1'
model = hub.load(module_handle)

image_url = 'https://cptstatic.s3.amazonaws.com/imagens/enviadas/materias/materia16043/caracteristicas-cavalos-saudaveis-artigos-cursos-cpt.jpg'
response = requests.get(image_url)
image = Image.open(io.BytesIO(response.content)).resize((224, 224))
```

```python
input_arr = np.array([keras.applications.mobilenet_v2.preprocess_input(np.array(image))])

input_tensor = tf.convert_to_tensor(input_arr, dtype=tf.float32)
```

```python
detector = model.signatures['default']
outputs = detector(input_tensor)

detection_boxes = outputs['detection_boxes']
detection_scores = outputs['detection_scores']
detection_class_labels = outputs['detection_class_labels']
detection_class_names = outputs['detection_class_names']
```

```python
plt.imshow(image)
for box, score, label in zip(detection_boxes[0], detection_scores[0], detection_class_names[0]):
    if score > 0.5:  # Definindo um limite de confiança para exibir as detecções
        ymin, xmin, ymax, xmax = box.numpy()
        left, right, top, bottom = xmin * 224, xmax * 224, ymin * 224, ymax * 224
        plt.Rectangle((left, top), right - left, bottom - top, fill=False, color='red')
        plt.text(left, top, label.decode('utf-8'), color='red')
plt.axis('off')
plt.show()
```
